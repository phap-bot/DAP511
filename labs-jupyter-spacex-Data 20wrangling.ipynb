```python
# Pandas là một thư viện phần mềm được viết cho ngôn ngữ lập trình Python để thao tác và phân tích dữ liệu.
import pandas as pd

# NumPy là một thư viện cho ngôn ngữ lập trình Python, hỗ trợ các mảng và ma trận đa chiều lớn, cùng với một tập hợp lớn các hàm toán học để thao tác trên các mảng này.
import numpy as np

# Tải bộ dữ liệu SpaceX từ URL
URL = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DS0321EN-SkillsNetwork/datasets/dataset_part_1.csv"
df = pd.read_csv(URL)

# Hiển thị 10 dòng đầu tiên của dữ liệu để xem qua
print("10 dòng dữ liệu đầu tiên:")
print(df.head(10))
print("\n" + "="*50 + "\n")

# Xác định và tính toán tỷ lệ phần trăm các giá trị bị thiếu trong mỗi thuộc tính (cột)
print("Tỷ lệ phần trăm giá trị bị thiếu mỗi cột:")
print(df.isnull().sum() / df.count() * 100)
print("\n" + "="*50 + "\n")

# Xác định các cột nào là số và các cột nào là phân loại (categorical)
print("Kiểu dữ liệu của mỗi cột:")
print(df.dtypes)
print("\n" + "="*50 + "\n")


#------------------------------------------------------------------



#------------------------------------------------------------------
# TASK 2: Tính số lượng và tần suất của mỗi quỹ đạo
#------------------------------------------------------------------
print("TASK 2: Số lượng và tần suất của mỗi quỹ đạo")
# Sử dụng phương thức .value_counts() để xác định số lượng và tần suất của mỗi quỹ đạo trong cột 'Orbit'
orbit_counts = df['Orbit'].value_counts()
print(orbit_counts)
print("\n" + "="*50 + "\n")


#------------------------------------------------------------------
# TASK 3: Tính số lượng và tần suất của các kết quả nhiệm vụ theo loại quỹ đạo
#------------------------------------------------------------------
print("TASK 3: Số lượng và tần suất của các kết quả hạ cánh")
# Sử dụng phương thức .value_counts() trên cột 'Outcome' để xác định số lượng kết quả hạ cánh.
# Sau đó gán nó vào một biến landing_outcomes.
landing_outcomes = df['Outcome'].value_counts()
print(landing_outcomes)
print("\n")

# In ra các loại kết quả
for i, outcome in enumerate(landing_outcomes.keys()):
    print(i, outcome)
print("\n")

# Chúng ta tạo một tập hợp các kết quả mà giai đoạn thứ hai không hạ cánh thành công
# Dựa trên danh sách in ở trên, các kết quả xấu là ở chỉ số 1, 3, 5, 6, 7
bad_outcomes = set(landing_outcomes.keys()[[1, 3, 5, 6, 7]])
print("Các kết quả hạ cánh không thành công (bad_outcomes):")
print(bad_outcomes)
print("\n" + "="*50 + "\n")


#------------------------------------------------------------------
# TASK 4: Tạo nhãn kết quả hạ cánh từ cột 'Outcome'
#------------------------------------------------------------------
print("TASK 4: Tạo nhãn kết quả hạ cánh (landing_class)")
# landing_class = 0 nếu là bad_outcome
# landing_class = 1 nếu ngược lại
landing_class = []
for outcome in df['Outcome']:
    if outcome in bad_outcomes:
        landing_class.append(0)
    else:
        landing_class.append(1)

# In 10 giá trị đầu tiên của danh sách vừa tạo để kiểm tra
print("10 giá trị đầu tiên của landing_class:", landing_class[:10])
print("\n" + "="*50 + "\n")
```

```python
# TASK 1: Tính số lần phóng tại mỗi địa điểm phóng
print("TASK 1: Số lần phóng tại mỗi địa điểm")
# Áp dụng hàm value_counts() trên cột LaunchSite
launch_site_counts = df['LaunchSite'].value_counts()
print(launch_site_counts)
print("\n" + "="*50 + "\n")
```

```python
# PHẦN 2: THỰC HIỆN DATA WRANGLING VÀ XUẤT FILE
print("Thực hiện Data Wrangling và xuất file")

# Thêm cột 'Class' mới vào DataFrame từ danh sách landing_class
df['Class'] = landing_class

# Hiển thị 8 dòng đầu tiên của cột 'Class' để kiểm tra
print("8 dòng đầu của cột 'Class' mới:")
print(df[['Class']].head(8))
print("\n")

# Hiển thị 5 dòng đầu tiên của toàn bộ DataFrame sau khi đã thêm cột mới
print("5 dòng đầu của DataFrame sau khi thêm cột 'Class':")
print(df.head(5))
print("\n")

# Xác định tỷ lệ thành công
success_rate = df["Class"].mean()
print(f"Tỷ lệ hạ cánh thành công: {success_rate:.2%}")
print("\n" + "="*50 + "\n")

# Xuất DataFrame đã được xử lý ra file CSV mới
print("Xuất dữ liệu ra file 'dataset_part_2.csv'...")
df.to_csv("dataset_part_2.csv", index=False)
print("Hoàn thành!")
```
